# Univariate Analysis

## Numerical

```python
sns.histplot(tips['total_bill'])
plt.title('Distribution of Total Bill')
```

```python
sns.histplot(tips['tip'])
plt.title('Distribution of Tips')
```

```python
sns.histplot(tips['size'])
plt.title('Distribution of Size')
```

```python
sns.boxplot(x=tips['tip'])
plt.title('Boxplot of Tips')   #it shows the mean value distribution to find the outliers
```

```python
sns.boxplot(x=tips['size'])
plt.title('Boxplot of Size')   #it shows the mean value distribution
```

```python
sns.boxplot(x=tips['total_bill'])
plt.title('Boxplot of Total Bill')
```

## Categorical Colums

```python
sns.countplot(x='day', data=tips)
plt.title('Categorical Observation of Day')
```

```python
sns.countplot(x='Gender', data=tips)
plt.title('Categorical Observation of Gender')
```

```python
sns.countplot(x='time', data=tips)
plt.title('Categorical Observation of TIme')
```

```python
sns.countplot(x='smoker', data=tips)
plt.title('Categorical Observation of Smoker')
```

### Numerical-numerical Relationship

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='tip', palette='coolwarm')
plt.title('Total Bill Vs Tips')
plt.show()
```

```python
sns.scatterplot(x='size', y='tip', data=tips, hue='tip', palette='coolwarm')
plt.title('Total Size Vs Tips')
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='size', data=tips, hue='size', palette='coolwarm')
plt.title('Total Bill Vs Size')
plt.show()
```

```python
cor=tips.corr(numeric_only=True)
```

```python
sns.heatmap(cor, annot=True, cmap='coolwarm')
plt.title('Correlation between numerical values')
```
