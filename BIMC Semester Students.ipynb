```python
nan_rows = data[data['Total_Students'].isna()]
print("Rows with NaN in Total_Students:")
print(nan_rows)
```

```python
mean_value = data['Total_Students'].mean()
data['Total_Students'].fillna(mean_value, inplace=True)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Course', y='Total_Students', hue='Batch', data=data, palette='tab10')
plt.title('Total Students Per Course: Batchwise')
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Batch', y='Total_Students', hue='Course', marker='o', data=data, palette='tab10')
plt.title('Trend of Current Students by Course', fontsize=16)
plt.xlabel('Batch Year', fontsize=14)
plt.ylabel('Total Students', fontsize=14)
plt.legend(title='Course')
```

```python
data['Drop Out'] = data['Total_Seats'] - data['Total_Students']
```

```python
data.columns
```

```python
data.head()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Batch', y='Drop Out', hue='Course', marker='o', data=data, palette='tab10')
plt.title('Trend of Dropout of Students by Course', fontsize=16)
plt.xlabel('Batch Year', fontsize=14)
plt.ylabel('Total Drop Out Students', fontsize=14)
plt.legend(title='Course')
```

```python
data['drop_percent']=(data['Drop Out']/data['Total_Seats'])*100
```
