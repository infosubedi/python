```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flights = sns.load_dataset("flights")
print(flights.head())
```

```python
flights.describe()
```

```python
flights.shape
```

```python
flights.head()
```

```python
flights.isnull().sum()
```

```python
flights.drop_duplicates()
```

```python
sns.histplot(flights['passengers'])
plt.title('Distribution of Passengers')
```

```python
sns.boxplot(x='month', y='passengers', data=flights)
plt.title('Boxplot of Month Vs Passengers')
```

```python
# Group by year and sum the passengers
total_passengers_per_year = flights.groupby('year')['passengers'].sum().reset_index()

# Display the result
print(total_passengers_per_year)
```

```python
total_passenger = flights.groupby('year')['passengers'].sum().reset_index()
```

```python
sns.barplot(x='year', y='passengers', data=total_passenger, palette='coolwarm')
plt.title('Total Passengers per Year')
```

```python
# Group by year and sum the passengers
total_passengers_per_month = flights.groupby('month')['passengers'].sum().reset_index()

# Display the result
print(total_passengers_per_month)
```

```python
total_passenger_m = flights.groupby('month')['passengers'].sum().reset_index()
```

```python
sns.barplot(x='month', y='passengers', data=total_passenger_m, palette='viridis')
plt.title('Total Passengers per Month')
```

```python
sns.lineplot(x='month', y='passengers', data=total_passenger_m, marker='o', color='b')
plt.title('Total Passengers per Year')
```
