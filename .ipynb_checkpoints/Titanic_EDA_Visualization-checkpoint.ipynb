```python
#required libraries
import warnings
import pandas as pd
import seaborn as sns   #visualization lib
import matplotlib.pyplot as plt

# Supress FutureWarnings
warnings.simplefilter(action='ignore', category=FutureWarning)

#Load Dataset
df=pd.read_csv('data/titanic.csv',index_col='PassengerId')
#Display first few rows
df.head()
```

```python
# Bar plot for Survived column
plt.figure(figsize=(8,4))
sns.countplot(x='Survived', data=df, hue='Survived', palette='Set3')
plt.title('Count of Survived vs Not Survived')
plt.xlabel('Survived (1 = Yes, 0 = No)')
plt.ylabel('Passenger Count')
plt.show()
```

```python
pclass_counts = df['Pclass'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(pclass_counts, labels=pclass_counts.index, autopct='%1.1f%%', colors=['yellow', 'red', 'green'], startangle=180)
plt.title('Passenger Class Distribution')
plt.show()
```
